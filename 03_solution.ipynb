```python
cols_needed = ['latitude',
'longitude',
'housing_median_age',
'total_rooms',
'total_bedrooms',
'population',
'households',
'median_income',
'median_house_value',
'ocean_proximity']

data = data[cols_needed]
data.head()
```

```python
data.isna().sum()
```

## total_bedrooms contain 207 null values

```python
data.fillna(value=0, inplace=True)
data['rooms_per_household'] = data.total_rooms / data.households
data['bedrooms_per_room'] = data.total_bedrooms / data.total_rooms    
data['population_per_household'] = data.population / data.households
data.head(5)
```

# Question1

```python
#frequent observation mode of ocean proximity
data['ocean_proximity'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

full_train, test = train_test_split(data, train_size=0.8, test_size=0.2, random_state=42, shuffle=True)
train, val = train_test_split(full_train, train_size=0.75, test_size=0.25, random_state=42, shuffle=True)

full_train.reset_index(drop=True)
train.reset_index(drop=True)
val.reset_index(drop=True)
test.reset_index(drop=True)
```

```python
print(data.shape)
print(full_train.shape)
print(train.shape)
print(val.shape)
print(test.shape)

train_x = train.drop('median_house_value', axis=1)
train_y = train[['median_house_value']]

val_x = val.drop('median_house_value', axis=1)
val_y = val[['median_house_value']]

test_x = test.drop('median_house_value', axis=1)
test_y = test[['median_house_value']]
```

# Question 2

```python
train_x.corr()
```

total_bedrooms and households have high correleations

```python
from IPython.display import display
full_train['above_average'] = full_train['median_house_value'] > full_train['median_house_value'].mean()
full_train['above_average'] = full_train['above_average'].astype(int)
display(full_train['above_average'].value_counts())

train_y['above_average'] = train_y['median_house_value'] > train_y['median_house_value'].mean()
train_y['above_average'] = train_y['above_average'].astype(int)
display(train_y['above_average'].value_counts())

val_y['above_average'] = val_y['median_house_value'] > val_y['median_house_value'].mean()
val_y['above_average'] = val_y['above_average'].astype(int)
display(val_y['above_average'].value_counts())

test_y['above_average'] = test_y['median_house_value'] > test_y['median_house_value'].mean()
test_y['above_average'] = test_y['above_average'].astype(int)
display(test_y['above_average'].value_counts())
```

# Question3
mutual informatin score betwen above average and ocean prodimity and round it 2

```python
from sklearn.metrics import mutual_info_score
mutual_info_score(train_y.above_average, train_x.ocean_proximity).round(2)
```

```python
from sklearn.feature_extraction import DictVectorizer

dv = DictVectorizer(sparse=False)
train_vect_x = dv.fit_transform(train_x.to_dict(orient='records'))
train_vect_x = pd.DataFrame(train_vect_x, columns=dv.get_feature_names_out())
display(train_vect_x.head(5))

val_vect_x = dv.transform(val_x.to_dict(orient='records'))
val_vect_x = pd.DataFrame(val_vect_x, columns=dv.get_feature_names_out())
display(val_vect_x.head(5))

test_vect_x = dv.transform(test_x.to_dict(orient='records'))
test_vect_x = pd.DataFrame(test_vect_x, columns=dv.get_feature_names_out())
display(test_vect_x.head(5))
```

# Question 4
Train and output the accuracy for validation dataset

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state=42)
model.fit(train_vect_x, train_y.above_average)

val_pred_y = model.predict_proba(val_vect_x)[:,1]
val_pred_y = val_pred_y >= 0.5
val_pred_y = val_pred_y.astype(int)

all_features_accuracy = (val_y.above_average == val_pred_y).mean()
all_features_accuracy
```

# Question 5
least feature among :
total_rooms
total_bedrooms
population
households

```python
for col in ['total_rooms', 'total_bedrooms', 'population', 'households']:
    model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state=42)

    model.fit(train_vect_x.drop(col, axis=1), train_y.above_average)

    val_pred_y = model.predict_proba(val_vect_x.drop(col, axis=1))[:,1]
    val_pred_y = val_pred_y >= 0.5
    val_pred_y = val_pred_y.astype(int)

    features_except_call_accuracy = (val_y.above_average == val_pred_y).mean()
    print(f'Impact of abscence of feature {col} in accuracy is : ', np.abs(all_features_accuracy - features_except_call_accuracy))
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error

train_y['median_house_value_log1p'] = np.log1p(train_y['median_house_value'])
val_y['median_house_value_log1p'] = np.log1p(val_y['median_house_value'])
test_y['median_house_value_log1p'] = np.log1p(test_y['median_house_value'])

for a in [0, 0.01, 0.1, 1, 10]:
    model = Ridge(alpha=a, solver="sag", random_state=42)
    model.fit(train_vect_x, train_y.median_house_value_log1p)

    val_pred_log_y = model.predict(val_vect_x)
    val_pred_y = np.expm1(val_pred_log_y)
    val_orig_y = np.expm1(val_y.median_house_value_log1p)
    result= mean_squared_error(val_y.median_house_value_log1p, val_pred_log_y, squared=False).round(3)
    print(f'RMSE of log median house value value with alpha {a} is : ', result)
    result= mean_squared_error(val_orig_y, val_pred_y, squared=False).round(3)
    print(f'RMSE of median house value value with alpha {a} is : ', result)
```
